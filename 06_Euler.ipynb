```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
%matplotlib inline

f = lambda t, s: np.exp(-t)
h = 0.1
t = np.arange(0, 1 + h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])


plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
%matplotlib inline

def f(t, s):
    return np.exp(-t)

def y_taylor_series(x):
    return 1 - 4*x + 8*x**2 - (31/3)*x**3 + (31/3)*x**4

def y_analytical(x):
    return (31/32)*np.exp(-4*x) + (1/4)*x**2 - (1/8)*x + (1/32)

x = 0.1
y_taylor = y_taylor_series(x)

y_exact = y_analytical(x)

estimated_error = abs(y_taylor - 1)

actual_error = abs(y_exact - y_taylor)

print(f"y(0.1) using Taylor series: {y_taylor}")
print(f"y(0.1) using analytical solution: {y_exact}")
print(f"Estimated error: {estimated_error}")
print(f"Actual error: {actual_error}")

h = 0.1
t = np.arange(0, 1 + h, h)
s0 = 1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

s_taylor = np.zeros(len(t))
s_taylor[0] = s0

for i in range(1, len(t)):
    s_taylor[i] = y_taylor_series(t[i])

s_exact = np.zeros(len(t))
for i in range(len(t)):
    s_exact[i] = y_analytical(t[i])

plt.figure(figsize=(12, 8))
plt.plot(t, s_taylor, 'bo--', label='Taylor Series Approximation')
plt.plot(t, s_exact, 'g', label='Exact Solution')
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
%matplotlib inline

# Define the differential equation y' = sin(y)
f = lambda t, s: np.sin(s)

# Step size
h = 0.1

# Time range from 0 to 0.5
t = np.arange(0, 0.5 + h, h)

# Initial condition y(0) = 1
s0 = 1

# Initialize the solution array
s = np.zeros(len(t))
s[0] = s0

# Euler's method to solve the differential equation
for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])
    s[i + 1] = np.round(s[i + 1], 4)  # Keep four decimal places in the computations

# Plotting the results
plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.title('Approximate Solution for y = sin(y) with y(0) = 1')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

# Print the results
for i in range(len(t)):
    print(f"t = {t[i]:.1f}, y(t) = {s[i]:.4f}")
```
